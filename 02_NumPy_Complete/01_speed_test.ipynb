```python
## USing Python List
import numpy as np
import time

size = 1_000_000

l1= list(range(size))
l2= list(range(size))

start = time.time()

add = [x+y for x,y in zip (l1,l2)]

end = time.time()

print(add[0:10])
print(start)
print(end)
print(end - start)
```

```python
## USing Numpy Arrays
import numpy as np
import time

size = 1_000_000

l1= np.array(list(range(size)))
l2= np.array(list(range(size)))

start = time.time()

add = l1 + l2
end = time.time()
print(add[0:10])

print(end - start)
```
